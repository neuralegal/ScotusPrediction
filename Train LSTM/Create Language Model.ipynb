#### Create Language Model: Train until accuracy begins decreasing then reload weights from best model

```python
from fastai.text import *
# Requires fast.ai 1.0.48

texts = '../Dataframes'
data_lm = (TextLMDataBunch.from_csv(texts,csv_name="train_language_model.csv",text_cols=0,valid_pct = 0.1))
learn = language_model_learner(data_lm, AWD_LSTM, drop_mult=0.3)

learn.fit_one_cycle(1, 1e-02, moms=(0.8,0.7))
learn.save('head') 
learn.load('head'); 

learn.unfreeze()
learn.fit_one_cycle(5,1e-3, moms=(0.8,0.7))
learn.save('5')
learn.fit_one_cycle(5,1e-3, moms=(0.8,0.7))
learn.save('10')
learn.fit_one_cycle(5,1e-3, moms=(0.8,0.7))
learn.save('15') 
learn.fit_one_cycle(5,1e-3, moms=(0.8,0.7))
learn.save('20') 
learn.fit_one_cycle(5,1e-3, moms=(0.8,0.7))
learn.save('25') 
learn.fit_one_cycle(5,1e-3, moms=(0.8,0.7))
learn.save('30') 
```

```python
learn.load("25") #Reload the best performing weights
learn.save('finalLM')
learn.save_encoder('lm_enc') #Encoder is @ 0.459
```
